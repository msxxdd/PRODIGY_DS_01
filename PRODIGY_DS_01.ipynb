```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Sample dataset
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve', 'Frank', 'Grace', 'Helen', 'Ian', 'Jack'],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Female', 'Male', 'Male'],
    'Age': [23, 35, 45, 29, 41, 33, 27, 30, 38, 40]
}

# Convert dataset to a DataFrame
df = pd.DataFrame(data)

# Bar chart for Gender distribution
plt.figure(figsize=(8, 5))
df['Gender'].value_counts().plot(kind='bar', color=['skyblue', 'pink'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Histogram for Age distribution
plt.figure(figsize=(8, 5))
plt.hist(df['Age'], bins=np.arange(20, 50, 5), color='orange', edgecolor='black', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
